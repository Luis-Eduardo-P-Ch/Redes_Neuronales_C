Programacion de una red neuronal:

    

```
# En primer lugar definanmos una clase, un tipo de objeto que se refiera a una capa.
#Clase de la capa de la red
class neural_layer():
  def __init__(self, n_conn, n_neur, act_f):

    self.act_f = act_f

    self.b = np.random.rand(1, n_neur) * 2 -1
    self.W = np.random.rand(n_conn, n_neur) * 2 -1
```

```
# Funciones de Activacion

sigm= (lambda x: 1/ (1 + np.e ** (-x)),
       lambda x: x * (1 - x))

relu = lambda x: np.maximum(0, x)

_x = np.linspace(-5, 4, 100)
plt.plot(_x, sigm[0](_x))
#plt.plot(_x, relu(_x))

```

# Una vez creada la clase que nos permite crear capas podemos empezar a diseñar una red neuronal

```
#l0 = neural_layer(p, 4, sigm)
#l1 = neural_layer(4, 8, sigm)

def create_nn(topology, act_f):
  nn = []
  for l, layer in enumerate(topology[:-1]):
    nn.append(neural_layer(topology[l], topology[l+1], act_f))
  return nn

topology = [p, 4, 8, 16, 8, 4, 1]

create_nn(topology, sigm)
```
